## Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12,4  # with x height (inches)
sns.set_style('whitegrid')
```

## Labeling plot features
### The Functional Method

plt.xlabel()

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]


# for bar chart
plt.bar(x,y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
# for pie chart

z = [1,2,3,4,0.5]
veh_type = ['bicycle', 'motorbike','car','van','stroller'] # labels
plt.pie(z, labels=veh_type) # data, labels
plt.show()
```

### The object-oriented method

```python
address = '/workspaces/LinkedinLearning_python-for-data-science-part-1/data/mtcars.csv'

# data
cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb'] # assign column names

mpg = cars.mpg # select the mpg column: 目的 to isoltae the mpg column and plot it

# containers
fig = plt.figure() # figure object
ax = fig.add_axes([.1,.1,1,1]) # axes object; left, bottom, width, height

# plot out the data
mpg.plot() 

# set ticks 设置x轴刻度
ax.set_xticks(range(32))

# set labels
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium') # tick_lable; rotation: 旋转角度
ax.set_title('Miles Per Gallon of Cars in mtcars') # title

ax.set_xlabel('car names') # x-axis label
ax.set_ylabel('miles/gal')
```

## Adding a legend to your plot
### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best') # best (location): 自动选择最佳位置
plt.show()
```

### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```

## Annotating your plot

```python
mpg.max() # 33.9 --max value of mpg_column
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32)) # 32个数据点

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.set_ylim([0,45]) # create a limit for y-axis --[min, max]

ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35),arrowprops=dict(facecolor='black', shrink=0.05))
# xy: the point you want to annotate; 
# xytext: the location of the text; 
# arrowprops: the arrow properties --construct a dictinary

ax.legend(loc='best') # legend: 图例
```
