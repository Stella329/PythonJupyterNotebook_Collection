## Defining axes, ticks, and grids

#### 笔记：Axe vs. Axis

Axes: An Axes object is a part of a figure where data is plotted. It contains methods for plotting data, setting labels, and customizing the plot. The Axes object includes the x-axis and y-axis, the ticks, the labels, the title, and the plot itself. In your code, ax is an Axes object.

Axis: An Axis object represents a single axis (either x or y) in the plot. It contains methods for setting the properties of the axis, such as the limits, ticks, and labels. 

The Axes object has two Axis objects: one for the x-axis and one for the y-axis.

```python
# set axes properties

x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure() # create a figure object -- a blank canvas
ax1 = fig.add_axes([.1,.1,1,1]) # axes object -- a part of the figure object; set its position and size
# ax2 = fig.add_axes([1,1,1,1]) # 我的vs.

ax1.plot(x,y) # add data
# ax2.plot(x,y)
```

```python
# 笔记

ax = fig.add_axes([.1,.1,1,1]) 

# fig.add_axes[left, bottom, width, height] ##--in relative coordinates (0 to 1).
# .1: The left position of the axes (10% from the left of the figure).
# .1: The bottom position of the axes (10% from the bottom of the figure).
# 1: The width of the axes (100% of the figure's width).
# 1: The height of the axes (100% of the figure's height).
# This means the axes will start 10% from the left and bottom of the figure and will span the entire width and height of the figure.
```

```python
# 笔记

# ax is an Axes object created with: set the limits for the x-axis and y-axis, respectively.
fig.add_axes([.1, .1, 1, 1]).
ax.set_xlim([1, 9])
ax.set_ylim([0, 5]) 

# Axes object set the tick marks for the x-axis and y-axis, respectively.
ax.set_xticks([0, 1, 2, 3, 4, 5, 6, 8, 9, 10]) 
ax.set_yticks([0, 1, 2, 3, 4, 5]) 
```

```python
# 改变x,y轴的刻度properties

x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

# set limit of the axes (--x, y)
ax.set_xlim([1,9])
ax.set_ylim([0,5])

# set tick marks
ax.set_xticks([0,1,2,3,4,5,6,8,9,10]) # 删除7th刻度
ax.set_yticks([0,1,2,3,4,5]) # Y轴 max_limit =5

ax.plot(x,y)
```

```python
# 背景加grid background

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

## Subplots: Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig,(ax1, ax2) = plt.subplots(1,2) # create a figure with 2 axes (to add 2 subplots) --tuple
ax1.plot(x)
ax2.plot(x,y)
```

### To create more than 2 subplots using plt.subplots(1,2)

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
# 2x2 grid of subplots --This will create four subplots arranged in two rows and two columns


fig, axes = plt.subplots(3, 3)
# In this case, axes will be a 2D array of subplot axes that you can iterate over or index into as needed.
```

```python
# MORE i.e. on fig, axes = plt.subplots(3, 3)

import matplotlib.pyplot as plt
import numpy as np

# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(10, 10)) # creates a 3x3 grid of subplots with a specified figure size

# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Loop through each subplot and plot the data
for i in range(3):
  for j in range(3):
    axes[i, j].plot(x, y)
    axes[i, j].set_title(f'Subplot {i+1},{j+1}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
