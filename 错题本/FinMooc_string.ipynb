## 题目：倒叙输出string 
End to Begining

```python
# Write your solution here
string = input('Please type in a string: ')
for item in range(1,len(string)+1):
    index = item * -1
    print(string[index])
```

```python
# Solution for part03-10_end_to_beginning
input_string = input("Please type in a string: ")
index = -1
while index >= -len(input_string):
    print(input_string[index])
    index -= 1
```

## 题目：A framed word

<Finland Mooc> Intro Part3 -> Topic 2


```python
# Write your solution here
word = input('Word:')
_1stline = '*'*30
space_len = 30-2-len(word)

space1 = space_len//2
if len(word)%2 !=0:
    space2 = space1+1
else:
    space2 = space1

print(_1stline)
print('*'+ space1*' ' + word + space2*' ' + '*' )
print(_1stline)
```

## 题目：Slice String -Reversed

```python
# Write your solution here

word = input('Please type in a string: ')
step = 1

while step <=len(word):
    print(word[-1*step:])
    step+=1
```

```python
# Solution for part03-18_substrings_part_2 --src/substrings_part_2.py

string = input("Please type in a string: ")
 
start = len(string) - 1
while start >= 0:
    print(string[start:])
    start -= 1
```
