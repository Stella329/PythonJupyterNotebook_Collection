# Concatenating data (/join)

```python
DF_obj = DataFrame(np.arange(36).reshape(6,6))  # np.arange(36) -> range(0, 35)
DF_obj

# np.arange(start, end, step=1)  -> return ndarray (--n-dimentional array, np核心数据结构)
# 类似python函数：range()  -> return range_obj
```

```python
DF_obj_2 = DataFrame(np.arange(15).reshape(5,3))
DF_obj_2
```

```python
pd.concat([DF_obj, DF_obj_2], axis=1)

# axis=1: concatonate/join based on row_index (--append columns)
```

```python
pd.concat([DF_obj, DF_obj_2])

# by default axis=0: append rows
```

# Transforming data
## Dropping data

```python
DF_obj.drop([0,2]) # drop the rows of which index=0 and =2  删除第0行和第2行
```

```python
DF_obj.drop([0,2], axis=1) # 删除第n列
```

## Adding data

### df.join()

adding a series_obj into a df_obj

```python
series_obj = Series(np.arange(6))
series_obj.name = "added_variable" # add a name to this series_obj
series_obj
```

```python
variable_added = DataFrame.join(DF_obj,series_obj)  # axis=0 by default: join by row_index --add columns
variable_added
```

### pd.concat()

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index=False) 
added_datatable

# concatenate A to A (to itself)
# ignore=False: 保留old index
```

```python
added_datatable = pd.concat([variable_added, variable_added], ignore_index=True) # 重新index (--regardless of original index)
added_datatable
```

## Sorting data
