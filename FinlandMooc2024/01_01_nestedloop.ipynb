```python
# Solution for part03-24_first_letters_of_words

sentence = input("Please type in a sentence: ")
 
# Add a space at the start, to make handling sentence easier
sentence = " " + sentence
 
# Searching for indexes which are preceded by spaces
index = 1
while index < len(sentence):
    if sentence[index-1] == " " and sentence[index] != " ": #查找空格后的第一个字母
        print(sentence[index])
    index += 1
```

## 题目：flip_the_pairs

Part3 -> section 3

```python
# Solution for part03-26_flip_the_pairs

number = int(input("Please type in a number: "))
 
index = 1
while index+1 <= number:
    print(index+1)
    print(index)
    index += 2
 
if index <= number:
    print(index)
```
